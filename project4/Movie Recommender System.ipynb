### Model

```python
matrix=movie_df.pivot_table(index='user_id', columns='title', values='rating')
matrix.head()
```

```python
x=matrix['Star Wars (1977)']
x
```

```python
x_corr= pd.DataFrame(matrix.corrwith(x), columns=['corr val'])
x_corr.dropna(inplace=True)
x_corr=pd.merge(x_corr,dataset, on='title')
x_corr
```

```python
x_corr=x_corr.sort_values('corr val', ascending=False)
x_corr
```

```python
x_corr=x_corr[x_corr['count']>=80]
x_corr.head()
```

```python
def recommend(col):
    x=matrix[col]
    x_corr= pd.DataFrame(matrix.corrwith(x), columns=['corr val'])
    x_corr.dropna(inplace=True)
    x_corr=pd.merge(x_corr,dataset, on='title')
    x_corr=x_corr.sort_values('corr val', ascending=False)
    x_corr=x_corr[x_corr['count']>=80]
    movies= x_corr.iloc[1:5,:1].values
    return movies.reshape(1,4)
```

```python
recommend('Star Wars (1977)')
```

```python
input_file=pd.read_csv('My_Ratings.csv')
input_file.drop(columns=input_file.columns[0],axis=1, inplace=True)
input_file
```

```python
p=0
q=1
for i in input_file['Movie Name']:
    input_file.iloc[p:q, 1:]=recommend(i)
    p+=1
    q+=1
input_file
```

```python
input_file.to_csv('My_Ratings.csv')
```
